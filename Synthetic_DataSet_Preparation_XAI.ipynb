```python
# importing pandas and numpy libraries
import numpy as np
import pandas as pd
```

## Creating Normal Data based on Domain Knowledge (Experience) -TYPE 1

```python
#[temp, humidity,aqi, noise, occupancy, co, co2, flame, data_sensitivity]

min_max = ([18, 30], [35, 65], [60, 80], [55, 70], [0, 2],[0,15],[600,1000],[0,1],[0,1])


def make_series(low, high, name):
    if any(isinstance(_, float) for _ in (low, high)):
        func = np.random.uniform
    else:
        func = np.random.randint
    return pd.Series(func(low, high, size=(333335,)), name=name)

column_names = ("temp", "humidity", "aqi", "noise", "occupancy", "co", "co2", "flame", "data_sensitivity")

df1 = pd.concat([make_series(lo, hi, name) for (lo, hi), name in zip(min_max, column_names)],
          axis=1)
```

```python
df1
```
